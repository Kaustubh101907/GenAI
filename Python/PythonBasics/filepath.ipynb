```python
## create a new directory

import os

new_directory = "my_new_directory"
os.mkdir(new_directory)
print(f"Directory '{new_directory}' created successfully.")
```

```python
##Listing all files and directories in the current directory
items = os.listdir('.')
print(f"Files and directories in the current directory: {items}")
```

```python
## joining paths

import os

dir_name = "my_new_directory"
file_name = "my_file.txt"
full_path = os.path.join(dir_name, file_name)
print(f"Full path: {full_path}")
```

```python
import os

dir_name = "my_new_directory"
file_name = "my_file.txt"
full_path = os.path.join(os.getcwd(), dir_name, file_name)
print(f"Full path: {full_path}")
```

```python
### getting the absolute path of a file

import os

relative_path = "my_file.txt"
absolute_path = os.path.abspath(relative_path)
print(f"Absolute path: {absolute_path}")
```
