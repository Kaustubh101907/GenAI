```python
## Leap year checker
year = int(input("Enter a year: "))
if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("this is a leap year")
        else:
            print("this is not a leap year")
    else:
        print("this is a leap year")
else:
    print("this is not a leap year")
```

```python
nested_tuple = ((1,2,3), ("a","b","c"), (True, False))

for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item, end=" ")
    print()  # for new line after each sub-tuple
```

```python
## Print all prime numbers between 1 and 100
for i in range(1, 101):
    if i > 1:
        for j in range(2, i):
            if (i % j) == 0:
                break
        else:
            print(i)
```

```python
numbers = [1,2,2,3,3,3,4,4,5,5,5,5,6,6,6]
frequency = {}
for number in numbers:
    if number in frequency:
        frequency[number] += 1
    else:
        frequency[number] = 1
print(frequency)
```
