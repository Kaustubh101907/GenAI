```python
## positional and keywords args

def print_details(*args, **kwargs):
    for arg in args:
        print(f"Positional argument: {arg}")
    for key, value in kwargs.items():
        print(f"Keyword argument: {key} = {value}") 
```

```python
def calculate_price(cart):
    total_price = 0
    for item in cart:
        total_price += item['price'] * item['quantity']
    return total_price


cart = [
    {'name': 'Apple', 'price': 0.5, 'quantity': 4},
    {'name': 'Banana', 'price': 0.3, 'quantity': 6},
    {'name': 'Orange', 'price': 0.7, 'quantity': 3}
]
total = calculate_price(cart)
print(f"Total price of the cart: ${total:.2f}")
```

```python
print_details("apple", "banana", fruit1="orange", fruit2="grape")
```

```python
def convert_temp(temp, unit):
    if unit == "C":
        return (temp * 9/5) + 32  # Convert Celsius to Fahrenheit
    elif unit == "F":
        return (temp - 32) * 5/9  # Convert Fahrenheit to Celsius
    else:
        raise ValueError("Unit must be 'C' or 'F'")
    
print(convert_temp(100, "C"))  # Output: 212.0
print(convert_temp(212, "F"))  # Output: 100.0
```

```python
def count_word_frequency(file_path):
    word_count = {}
    with open(file_path, 'r') as file:
        for line in file:
            words = line.split()
            for word in words:
                word = word.lower().strip('.,!?";()[]{}')  # Normalize the word
                if word:
                    word_count[word] = word_count.get(word, 0) + 1
    return word_count
```
