```python
### wordnet lemmatizer
from nltk.stem import WordNetLemmatizer
import nltk
lemmatizer = WordNetLemmatizer()
nltk.download('wordnet', quiet=True)
lemmatizer.lemmatize("running", pos="v")


words = {"eating", "eats", "history", "eaten", "writes", "wrote", "written", "goes", "went", "gone", "programming", "programs", "programmed", "finally", "final", "finals", "finalized",};

for word in words:
    print(word+" --> "+lemmatizer.lemmatize(word, pos="v"))
```
