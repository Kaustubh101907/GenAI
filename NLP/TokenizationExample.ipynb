```python
corpus = """Hello, my name is ChatGPT. Welcome to ChatGPT's era. I am an AI language model developed by OpenAI. I can assist you with a variety of tasks, including answering questions, providing explanations, generating text, and more. I am really smart! How can I help you today?"""

### Tokenization
## para --> sentence

from nltk.tokenize import sent_tokenize
documents =  sent_tokenize(corpus)
print(documents)
print(len(documents))


###for sentence in documents:
    ###print(sentence)'''

## tokenisation para into words or sentence into words

from nltk.tokenize import word_tokenize
words = word_tokenize(corpus)
print(words)


from nltk.tokenize import wordpunct_tokenize
words2 = wordpunct_tokenize(corpus)
print(words2)

from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
words3 = tokenizer.tokenize(corpus)
print(words3)
```
