```python
## clasification problem
## comments of a product is a positive review or negative review
## go goes gone == go

words = {"eating", "eats", "history", "eaten", "writes", "wrote", "written", "goes", "went", "gone", "programming", "programs", "programmed", "finally", "final", "finals", "finalized",};


## porter stemmer
from nltk.stem import PorterStemmer
porter_stemmer = PorterStemmer()
for word in words:
    print(word+" --> "+porter_stemmer.stem(word))
```

```python
## regexpstemmer class

from nltk.stem import RegexpStemmer
regexp_stemmer = RegexpStemmer('ing$|s$|e$|ed$', min=4)
regexp_stemmer.stem('eating')
```

```python
### Snowball stemmer class

words = {"eating", "eats", "history", "eaten", "writes", "wrote", "written", "goes", "went", "gone", "programming", "programs", "programmed", "finally", "final", "finals", "finalized",};

from nltk.stem import SnowballStemmer
snowball_stemmer = SnowballStemmer('english')
for word in words:
    print(word+" --> "+snowball_stemmer.stem(word))
```
